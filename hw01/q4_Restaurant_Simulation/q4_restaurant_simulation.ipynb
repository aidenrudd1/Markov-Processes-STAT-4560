## Problem 4b. Jill's Restaurant Simulation 

To start this problem, it is important to understand what is being asked. We will be simulating N times, the event that which 3 iid uniform random variables, $X_1, X_2, X_3$ ~ Uniform(0, 1). All three have arrived to the restaurant when the larges or the max of the X random variables has arrived. 

```python
# Theoretical pdf from part a. 
def f(t):
    return 3 * t ** 2

```

```python
import numpy as np

rng = np.random.default_rng(1)
list_tri = []
for i in range(3):
    list_tri.append(float(rng.random(1)))
print(list_tri)
set_max = max(list_tri)
print(set_max)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def main(seed: int = 0):
    rng = np.random.default_rng(seed)
    n = 100000
    Ts = []
    for i in range(n):
        list_tri = []
        for k in range(3):
            list_tri.append(rng.random())
        Ts.append(max(list_tri))
        
    x = np.linspace(0, 1, 100000)
    y = 3 * x ** 2
    plt.hist(Ts, bins = 100, density = True, color = "skyblue", edgecolor = "black")
    plt.plot(x, y, color = "red")
    plt.xlabel("t from [0, 1]")
    plt.ylabel("Estimated PDF")
    plt.title("Simulation of Restaurant Arrival")
    plt.tight_layout()
    plt.show()
```

```python
main(1)
```
